**loading data set - parkinsons disease prediction dataset**

```python
import pandas as pd
# load the local dataset (adjusted path)
data = pd.read_csv('dataset/Parkinsson disease.csv')
data
```

**dropping name column**

```python
# drop name column if it exists (avoid KeyError)
data = data.drop(columns=['name'], errors='ignore')
data
```

**features(x) and target(y) variables**

```python
x= data.drop(columns=['status'])
y = data['status']
y
```

**normalizing the data**

```python
# correct min-max normalization and reassign to x
nd = (x - x.min()) / (x.max() - x.min())
x = nd
nd
```

**splitting data into test and train**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=44)
```

**logistic regression**

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
yp_test=model.predict(x_test)
yp_train=model.predict(x_train)
```

```python
print(model.coef_)
```

**accuracy score,precision, recall,confusion matrix*,svm**

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,yp_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,yp_test))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,yp_test))
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(x_train, y_train)
svc
```

```python
yp = svc.predict(x_test)
print(accuracy_score(y_test, yp))
```

```python
print(classification_report(y_test,yp))
```

**KNN**

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
x_iris, y_iris = iris.data, iris.target
```

```python
# use separate variable names so we don't overwrite Parkinsons data/splits
x_iris_train, x_iris_test, y_iris_train, y_iris_test = train_test_split(x_iris, y_iris, test_size=0.3, random_state=42)
```

```python
k = 3
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(x_iris_train, y_iris_train)
```

```python
y_pred = knn.predict(x_iris_test)
```

```python
print("Accuracy:", accuracy_score(y_iris_test, y_pred))
print("\nClassification Report:\n", classification_report(y_iris_test, y_pred))
```

**bootstrapping**

```python
import numpy as np
import pandas as pd

def bootstrap(data, n_iterations=1000, statistic=np.mean, random_seed=42):
    np.random.seed(random_seed)
    bootstrap_stats = []
    for _ in range(n_iterations):
        sample = np.random.choice(data, size=len(data), replace=True)
        bootstrap_stats.append(statistic(sample))
    return np.array(bootstrap_stats)
if 'MDVP:Fo(Hz)' in data.columns:
    mdvp_fo_column = data['MDVP:Fo(Hz)']
    n_iterations = 1000

    bootstrap_means = bootstrap(mdvp_fo_column, n_iterations=n_iterations, statistic=np.mean)

    confidence_level = 0.95
    lower_bound = np.percentile(bootstrap_means, (1 - confidence_level) / 2 * 100)
    upper_bound = np.percentile(bootstrap_means, (1 + confidence_level) / 2 * 100)
    print(f"Bootstrap mean estimate: {np.mean(bootstrap_means):.4f}")
    print(f"{confidence_level*100:.1f}% confidence interval: ({lower_bound:.4f}, {upper_bound:.4f})")
else:
    print("The dataset does not contain the 'MDVP:Fo(Hz)' column. Please check the dataset structure.")
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
pca = PCA(n_components=None)
x_pca = pca.fit_transform(x_scaled)
explained_variance = pca.explained_variance_ratio_
```

```python
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(explained_variance) + 1), explained_variance.cumsum(), marker='o', linestyle='--')
plt.title('Cumulative Explained Variance by PCA Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid()
plt.show()
```

```python
print("Explained Variance Ratio by Component:")
for i, ratio in enumerate(explained_variance, start=1):
    print(f"Component {i}: {ratio:.4f}")
```
